# Problem Statement 1:Smart Parking Lot Management System

```python
def manage_parking(parking_capacity,vehicle_logs):

    parked_count = 0

    peak_usage = 0

    for log in  vehicle_logs:
        if log == "IN":
            parked_count +=1
        elif log =="OUT":
            if parked_count >0:
                parked_count -=1

    if parked_count > peak_usage:
         print("Peak Parking Usage:", peak_usage)

    if parked_count < parking_capacity :
         status ="Available"
    else:
         status ="Unavailable"
    return parked_count,status

parking_capacity=50
vehicle_logs=["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parked_count,status = manage_parking(parking_capacity,vehicle_logs)

print("Currently Parked Vehicles:", parked_count)
print("Parking Status: ",status)
```

# Problem Statement 2:Online Food Delivery Time Estimator

```python

#Create a function that estimates delivery time based on:
#Distance (km)
#Weather condition
#Traffic level
#Apply delays dynamically and display final ETA.

def delivery_time (Distance,Weather,Traffic_level):

    base_time = Distance * 5

    if Traffic_level =="Low":
        base_time += 5
    elif Traffic_level =="Medium":
        base_time += 10
    elif Traffic_level =="High":
        base_time += 15

    if Weather =="clear":
        base_time += 0
    elif Weather =="Cloudy":
        base_time += 10
    elif Weather =="Rainy":
        base_time += 20

    return base_time


Distance = 8
Traffic_level="High"
Weather= "Rainy"


base_time=delivery_time (Distance,Weather,Traffic_level)

print("Estimated Delivery Time:",base_time,"minutes")
```

#Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python

#Build a function that analyzes seat booking data and:
#Calculates occupancy percentage
#Determines if show is Housefull
#Suggests opening additional shows

def seat_occupancy_analyzer(total_seats, booked_seats):

    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        Status= "Housefull"
    elif occupancy >= 75:
        Status="Almost Full"
    elif occupancy >= 40:
        Status= "Filling Fast"
    else:
        Status="Seats Available"

    return occupancy ,Status

total_seats = 200
booked_seats = [1] * 150

occupancy, Status =seat_occupancy_analyzer(total_seats, booked_seats)

print("Occupancy:",occupancy, "%")
print("Show Status:  ",Status)
```

#Problem Statement 4:Cloud Server Load Classification System

```python

#Create a function to classify server load based on CPU usage readings.
#Average CPU < 50% → Normal
#50%–80% → Warning
#80% → Critical

def cpu_server_load(cpu_readings):

    total = 0

    # Loop to calculate sum
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    # Classification
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return average, status


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

average, status = cpu_server_load(cpu_readings)

print("Average CPU Load:", average, "%")
print("Server Status:", status)
```

#Problem Statement 5:Smart Classroom Resource Usage Monitor

```python

def resource_usage_monitor(resource_usage):

    overused = []

    # Loop through dictionary
    for resource, hours in resource_usage.items():

        if hours > 8:
            overused.append(resource)

    # Determine energy alert
    if overused:
        alert = "Yes"
    else:
        alert = "No"

    return overused, alert


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused_resources, energy_alert = resource_usage_monitor(resource_usage)

print("Overused Resources:", ", ".join(overused_resources))
print("Energy Alert:", energy_alert)
```

#Problem Statement 6: Online Event Registration Capacity Controller

```python

#Create a function that manages event registrations by:
#Tracking registrations
#Preventing overbooking
#Triggering waitlist mode

def registration_controller(capacity, registrations):

    confirmed = 0
    waitlisted = 0

    # Loop through each registration
    for user in range(registrations):

        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Determine registration status
    if waitlisted > 0:
        status = "Closed"
    else:
        status = "Open"

    return confirmed, waitlisted, status


# Sample Input
capacity = 100
registrations = 105

confirmed, waitlisted, status = registration_controller(capacity, registrations)

print("Confirmed Registrations:", confirmed)
print("Waitlisted Users:", waitlisted)
print("Registration Status:", status)
```
