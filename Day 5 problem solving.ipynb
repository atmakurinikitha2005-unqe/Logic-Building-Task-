# Problem 1: Social Media – Post Engagement Analyzer

```python

#  Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    print("Total Likes:", total_likes)

    # Condition to check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


#  Usage
likes = [200, 150, 300, 400]   # You can change values
analyze_post_engagement(likes)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python

#  Agriculture – Rainfall Adequacy Checker

def check_rainfall(rainfall_list, required_level):
    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)

    print("Average Rainfall:", int(average_rainfall))

    # Conditional check
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


#  Usage
rainfall_data = [70, 75, 80, 65, 70]
required = 70

check_rainfall(rainfall_data, required)
```

# Problem 4: Social Media – Duplicate Account Detection

```python

#  Social Media – Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    unique_usernames = set(usernames)

    # Check if duplicates exist
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


#  Usage
user_list = ["nikil", "rahul", "ananya", "rahul"]
detect_duplicate_accounts(user_list)
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
# Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices):
    premium_crops = []

    # Loop to filter prices
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


#  Usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

# Problem 9: E-Learning – Student Attendance Eligibility System

```python

#  E-Learning – Student Attendance Eligibility System

def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0

    # Loop to count attended classes (1 = Present, 0 = Absent)
    for status in attendance_list:
        if status == 1:
            attended_classes += 1

    # Calculate attendance percentage
    percentage = (attended_classes / total_classes) * 100

    print("Attendance Percentage:", percentage)

    # Check eligibility
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Usage
attendance_data = [1, 1, 0, 1, 1]   # 4 out of 5 classes attended
check_attendance(attendance_data)
```

# Problem 10: Smart Electricity Bill Analyzer

```python

# Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    # Billing calculation using slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


# Usage
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Status:", status)
```
