# 1 .Unique Words in a Sentence

```python
sentence=input ("enter the Sentence:")
words=sentence.split()
unique_words=set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# 2 . Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
name = max(employees, key=employees.get)
salary = employees[name]

print(f"Highest Salary: {name} - {salary}")
```

# 3.Find Maximum and Minimum Value

```python
numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]

# Loop through list to find max and min
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
# Display results
print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

# 4.Count Products Above a Price Threshold

```python

# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Counter
count = 0

# Loop to check prices above 1000
for price in prices:
    if price > 1000:
        count += 1

# Display result
print("Products above 1000:", count)
```

# 5.Calculate Attendance Percentage

```python
# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count number of presents
present_count = attendance.count("P")

# the Total days attended
total_days = len(attendance)

#  used Calculate percentage
attendance_percentage = (present_count / total_days) * 100

#  to Display result
print("Attendance Percentage:", attendance_percentage)
```

# 7.Count Character Frequency

```python
# Input string
text = "python"

# Dictionary to store character frequency
char_count = {}

# Loop through each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

# Display result
print(char_count)
```

# 9.Checking if a Key Exists in a Dictionary

```python
# Dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee not found")
```

# 10.Calculate Average Marks

```python
# List of marks
marks = [70, 75, 80, 65]

# Calculate average
average = sum(marks) / len(marks)

# Display result
print("Average Marks:", average)
```
